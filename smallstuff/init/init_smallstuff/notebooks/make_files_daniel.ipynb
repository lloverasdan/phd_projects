```python
### Set up
file_name = '/p/work1/lloveras/adj_4km/input_48_96/bw_jul5_48h.nc'
netcdf_type = 'NETCDF3_64BIT_OFFSET'
ncfile = nc.Dataset(file_name,'w',format=netcdf_type)
wrf_file = nc.Dataset('/p/work1/lloveras/adj_4km/out_files/wrfout_ctl')

ti = 8
dx = 4. # km
cent_length = 2500. # km
coarse_fac = 8
ny_4km = 1800
nx_4km = 2000
nz_4km = 100
```
