```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import warnings
warnings.filterwarnings("ignore")
import matplotlib.transforms as mtransforms
from wrf import smooth2d

G = 9.81
dx = 4
rol = -1000

slp_ctl_48h = smooth2d(np.load('/p/work1/lloveras/adj_4km/processed/2d_fields/data_48h/slp_ctl_48h.npy'),4)

slp_adj_48h = smooth2d(np.load('/p/work1/lloveras/adj_4km/processed/2d_fields/data_48h/slp_adj_48h.npy'),4)
slp_wave_48h = smooth2d(np.load('/p/work1/lloveras/adj_4km/processed/2d_fields/data_48h/slp_wave_48h.npy'),4)

slp_adj10_48h = smooth2d(np.load('/p/work1/lloveras/adj_4km/processed/2d_fields/data_48h/slp_adj10_48h.npy'),4)
slp_wave10_48h = smooth2d(np.load('/p/work1/lloveras/adj_4km/processed/2d_fields/data_48h/slp_wave10_48h.npy'),4)

slp_adj100_48h = smooth2d(np.load('/p/work1/lloveras/adj_4km/processed/2d_fields/data_48h/slp_adj100_48h.npy'),4)
slp_wave100_48h = smooth2d(np.load('/p/work1/lloveras/adj_4km/processed/2d_fields/data_48h/slp_wave100_48h.npy'),4)

dslp_adj_48h = slp_adj_48h - slp_ctl_48h
dslp_wave_48h = slp_wave_48h - slp_ctl_48h

dslp_adj10_48h = (slp_adj10_48h - slp_ctl_48h)*10.
dslp_wave10_48h = (slp_wave10_48h - slp_ctl_48h)*10.

dslp_adj100_48h = (slp_adj100_48h - slp_ctl_48h)*100.
dslp_wave100_48h = (slp_wave100_48h - slp_ctl_48h)*100.
```

```python
x1_48 = 750 #775 #925 #775
x2_48 = 1126 #1151 #1426 #1151
y1_48 = 750 #750 #725 #750
y2_48 = 1126 #1126 #1226 #1126

yax1_48 = np.arange(y1_48*dx,y2_48*dx,500).astype(int)
yax2_48 = (yax1_48 - y1_48*dx)/float(dx)
yax2_48 = yax2_48.astype(int)
yax1_48 = ["%0.0f" % number for number in yax1_48]
xax1_48 = np.arange(x1_48*dx,x2_48*dx,500).astype(int)
xax2_48 = (xax1_48 - x1_48*dx)/float(dx)
xax2_48 = xax2_48.astype(int)
xax1_48 = ["%0.0f" % number for number in xax1_48]

slp_ctl_48h = np.roll(slp_ctl_48h,rol,axis=-1)[y1_48:y2_48,x1_48:x2_48]

dslp_adj_48h = np.roll(dslp_adj_48h,rol,axis=-1)[y1_48:y2_48,x1_48:x2_48]
dslp_wave_48h = np.roll(dslp_wave_48h,rol,axis=-1)[y1_48:y2_48,x1_48:x2_48]
dslp_adj10_48h = np.roll(dslp_adj10_48h,rol,axis=-1)[y1_48:y2_48,x1_48:x2_48]
dslp_wave10_48h = np.roll(dslp_wave10_48h,rol,axis=-1)[y1_48:y2_48,x1_48:x2_48]
dslp_adj100_48h = np.roll(dslp_adj100_48h,rol,axis=-1)[y1_48:y2_48,x1_48:x2_48]
dslp_wave100_48h = np.roll(dslp_wave100_48h,rol,axis=-1)[y1_48:y2_48,x1_48:x2_48]
```

```python
fig, axd = plt.subplot_mosaic([['top left','top center','top right','cbar1'],
                               ['bottom left','bottom center','bottom right','cbar2']],
                              constrained_layout=True, figsize=(16.5,9), 
                              gridspec_kw={'width_ratios':[1,1,1,0.1],'height_ratios':[1,1]})

trans = mtransforms.ScaledTranslation(1/25, -1/25, fig.dpi_scale_trans)

lvls = np.asarray([-9,-7,-5,-3,-1,1,3,5,7,9])*2.5
lvls2 = np.asarray([-16,-8,-4,-2,-0.5,0.5,2,4,8,16])
nm = colors.BoundaryNorm(lvls,256)
nm2 = colors.BoundaryNorm(lvls2,256)

clrs = ['cyan',#'#0000FF',
        '#3434FF',
        '#6666FF',
        '#9898FF',
        '#CCCCFF',
        '#FFFEFE',
        '#FFCCCC',
        '#FF9A9A',
        '#FF6666',
        '#FF3232',
        'yellow'] #'FF0000'
cmap = matplotlib.colors.LinearSegmentedColormap.from_list("",clrs)

#cmap = get_cmap('bwr')
cmap.set_under('darkgreen')
cmap.set_over('brown')

cs1 = axd['top left'].contour(slp_ctl_48h,colors='k',levels=np.arange(950,1060,4),linewidths=2,alpha=0.8)
axd['top left'].clabel(cs1,fmt='%1.0f',inline=1,levels=np.arange(950,1060,8),fontsize=10,colors='k')
im1 = axd['top left'].contourf(dslp_adj_48h,levels=lvls,extend='both',cmap=cmap,norm=nm,alpha=0.9)
axd['top left'].set_xticks(xax2_48)
axd['top left'].set_xticklabels(['','','',''])
axd['top left'].set_ylabel('South-North (km)',labelpad=6,fontsize=20)
axd['top left'].set_yticks(yax2_48)
axd['top left'].set_yticklabels(yax1_48)
axd['top left'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['top left'].set_aspect('equal')
axd['top left'].set_title('Full',pad=12,fontsize=24)
axd['top left'].grid()
axd['top left'].text(-0.575, 0.625, 'Adjoint',transform=axd['top left'].transAxes + trans,
            fontsize=24, verticalalignment='top',rotation='vertical')
axd['top left'].text(0.001, 1.0, '(a)',transform=axd['top left'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2),zorder=10)

cs2 = axd['top center'].contour(slp_ctl_48h,colors='k',levels=np.arange(950,1060,4),linewidths=2,alpha=0.8)
axd['top center'].clabel(cs2,fmt='%1.0f',inline=1,levels=np.arange(950,1060,8),fontsize=10,colors='k')
im2 = axd['top center'].contourf(dslp_adj10_48h,levels=lvls,extend='both',cmap=cmap,norm=nm,alpha=0.9)
axd['top center'].set_xticks(xax2_48)
axd['top center'].set_xticklabels(['','','',''])
axd['top center'].set_yticks(yax2_48)
axd['top center'].set_yticklabels(['','','',''])
axd['top center'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['top center'].set_aspect('equal')
axd['top center'].set_title('Rescaled 1/10',pad=12,fontsize=24)
axd['top center'].grid()
axd['top center'].text(0.001, 1.0, '(b)',transform=axd['top center'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2),zorder=10)

cs3 = axd['top right'].contour(slp_ctl_48h,colors='k',levels=np.arange(950,1060,4),linewidths=2,alpha=0.8)
axd['top right'].clabel(cs3,fmt='%1.0f',inline=1,levels=np.arange(950,1060,8),fontsize=10,colors='k')
im3 = axd['top right'].contourf(dslp_adj100_48h,levels=lvls,extend='both',cmap=cmap,norm=nm,alpha=0.9)
axd['top right'].set_xticks(xax2_48)
axd['top right'].set_xticklabels(['','','',''])
axd['top right'].set_yticks(yax2_48)
axd['top right'].set_yticklabels(['','','',''])
axd['top right'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['top right'].set_aspect('equal')
axd['top right'].set_title('Rescaled 1/100',pad=12,fontsize=24)
axd['top right'].grid()
axd['top right'].text(0.001, 1.0, '(c)',transform=axd['top right'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2),zorder=10)

cs4 = axd['bottom left'].contour(slp_ctl_48h,colors='k',levels=np.arange(950,1060,4),linewidths=2,alpha=0.8)
axd['bottom left'].clabel(cs4,fmt='%1.0f',inline=1,levels=np.arange(950,1060,8),fontsize=10,colors='k')
im4 = axd['bottom left'].contourf(dslp_wave_48h,levels=lvls2,extend='both',cmap=cmap,norm=nm2,alpha=0.9)
axd['bottom left'].set_xlabel('West-East (km)',labelpad=6,fontsize=20)
axd['bottom left'].set_xticks(xax2_48)
axd['bottom left'].set_xticklabels(xax1_48)
axd['bottom left'].set_ylabel('South-North (km)',labelpad=6,fontsize=20)
axd['bottom left'].set_yticks(yax2_48)
axd['bottom left'].set_yticklabels(yax1_48)
axd['bottom left'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['bottom left'].set_aspect('equal')
axd['bottom left'].grid()
axd['bottom left'].text(-0.575, 0.625, 'Wave',transform=axd['bottom left'].transAxes + trans,
            fontsize=24, verticalalignment='top',rotation='vertical')
axd['bottom left'].text(0.001, 1.0, '(d)',transform=axd['bottom left'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2),zorder=10)

cs5 = axd['bottom center'].contour(slp_ctl_48h,colors='k',levels=np.arange(950,1060,4),linewidths=2,alpha=0.8)
axd['bottom center'].clabel(cs5,fmt='%1.0f',inline=1,levels=np.arange(950,1060,8),fontsize=10,colors='k')
im5 = axd['bottom center'].contourf(dslp_wave10_48h,levels=lvls2,extend='both',cmap=cmap,norm=nm2,alpha=0.9)
axd['bottom center'].set_xlabel('West-East (km)',labelpad=6,fontsize=20)
axd['bottom center'].set_xticks(xax2_48)
axd['bottom center'].set_xticklabels(xax1_48)
axd['bottom center'].set_yticks(yax2_48)
axd['bottom center'].set_yticklabels(['','','',''])
axd['bottom center'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['bottom center'].set_aspect('equal')
axd['bottom center'].grid()
axd['bottom center'].text(0.001, 1.0, '(e)',transform=axd['bottom center'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2),zorder=10)

cs6 = axd['bottom right'].contour(slp_ctl_48h,colors='k',levels=np.arange(950,1060,4),linewidths=2,alpha=0.8)
axd['bottom right'].clabel(cs6,fmt='%1.0f',inline=1,levels=np.arange(950,1060,8),fontsize=10,colors='k')
im6 = axd['bottom right'].contourf(dslp_wave100_48h,levels=lvls2,extend='both',cmap=cmap,norm=nm2,alpha=0.9)
axd['bottom right'].set_xlabel('West-East (km)',labelpad=6,fontsize=20)
axd['bottom right'].set_xticks(xax2_48)
axd['bottom right'].set_xticklabels(xax1_48)
axd['bottom right'].set_yticks(yax2_48)
axd['bottom right'].set_yticklabels(['','','',''])
axd['bottom right'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['bottom right'].set_aspect('equal')
axd['bottom right'].grid()
axd['bottom right'].text(0.001, 1.0, '(f)',transform=axd['bottom right'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2),zorder=10)

cbar = fig.colorbar(im1, orientation="vertical", cax=axd['cbar1'],ticks=lvls)
cbar.ax.tick_params(labelsize=20, pad=10)

cbar2 = fig.colorbar(im5, orientation="vertical", cax=axd['cbar2'],ticks=lvls2)
cbar2.ax.tick_params(labelsize=20, pad=10)

plt.savefig('/p/work1/lloveras/adj_4km/figs_june/dslp_48h.pdf',bbox_inches='tight')
```
