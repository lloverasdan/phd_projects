```python
%pylab inline
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from wrf import smooth2d
import warnings
warnings.filterwarnings("ignore")
import matplotlib.transforms as mtransforms

G = 9.81
dx = 4
rol = -1000

dbz_ctl = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/dbz/mdbz_ctl_48h.npy')
dbz_adj = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/dbz/mdbz_adj_48h.npy')
dbz_wave = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/dbz/mdbz_wave_48h.npy')

v900_ctl = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/vars_900/v900_ctl.npy')[2]
v900_adj = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/vars_900/v900_adj.npy')[2]
v900_wave = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/vars_900/v900_wave.npy')[2]

z900_ctl = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/vars_900/z900_ctl.npy')[2]
z900_adj = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/vars_900/z900_adj.npy')[2]
z900_wave = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/vars_900/z900_wave.npy')[2]

v500_ctl = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/vars_500/v500_ctl.npy')[6]
v500_adj = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/vars_500/v500_adj.npy')[6]
v500_wave = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/vars_500/v500_wave.npy')[6]

z500_ctl = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/vars_500/z500_ctl.npy')[6]
z500_adj = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/vars_500/z500_adj.npy')[6]
z500_wave = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/vars_500/z500_wave.npy')[6]

v350_ctl = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/vars_350/v350_ctl.npy')[2]
v350_adj = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/vars_350/v350_adj.npy')[2]
v350_wave = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/vars_350/v350_wave.npy')[2]

z350_ctl = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/vars_350/z350_ctl.npy')[2]
z350_adj = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/vars_350/z350_adj.npy')[2]
z350_wave = np.load('/p/work1/lloveras/adj_4km/processed_data_revisions/vars_350/z350_wave.npy')[2]
```

```python
x1 = 650 #800 
x2 = 1401 #1301 
y1 = 600 #725
y2 = 1351 #1226 

yax1 = np.arange(y1*dx,y2*dx,1000).astype(int)
yax2 = (yax1 - y1*dx)/float(dx)
yax2 = yax2.astype(int)
yax1 = ["%0.0f" % number for number in yax1]
xax1 = np.arange(x1*dx,x2*dx,1000).astype(int)
xax2 = (xax1 - x1*dx)/float(dx)
xax2 = xax2.astype(int)
xax1 = ["%0.0f" % number for number in xax1]

dbz_ctl = np.roll(dbz_ctl,rol,axis=-1)[y1:y2,x1:x2]
dbz_adj = np.roll(dbz_adj,rol,axis=-1)[y1:y2,x1:x2]
dbz_wave = np.roll(dbz_wave,rol,axis=-1)[y1:y2,x1:x2]

v900_ctl = np.roll(v900_ctl,rol,axis=-1)[y1:y2,x1:x2]
v900_adj = np.roll(v900_adj,rol,axis=-1)[y1:y2,x1:x2]
v900_wave = np.roll(v900_wave,rol,axis=-1)[y1:y2,x1:x2]

z900_ctl = np.roll(z900_ctl,rol,axis=-1)[y1:y2,x1:x2]
z900_adj = np.roll(z900_adj,rol,axis=-1)[y1:y2,x1:x2]
z900_wave = np.roll(z900_wave,rol,axis=-1)[y1:y2,x1:x2]

v500_ctl = np.roll(v500_ctl,rol,axis=-1)[y1:y2,x1:x2]
v500_adj = np.roll(v500_adj,rol,axis=-1)[y1:y2,x1:x2]
v500_wave = np.roll(v500_wave,rol,axis=-1)[y1:y2,x1:x2]

z500_ctl = np.roll(z500_ctl,rol,axis=-1)[y1:y2,x1:x2]
z500_adj = np.roll(z500_adj,rol,axis=-1)[y1:y2,x1:x2]
z500_wave = np.roll(z500_wave,rol,axis=-1)[y1:y2,x1:x2]

v350_ctl = np.roll(v350_ctl,rol,axis=-1)[y1:y2,x1:x2]
v350_adj = np.roll(v350_adj,rol,axis=-1)[y1:y2,x1:x2]
v350_wave = np.roll(v350_wave,rol,axis=-1)[y1:y2,x1:x2]

z350_ctl = np.roll(z350_ctl,rol,axis=-1)[y1:y2,x1:x2]
z350_adj = np.roll(z350_adj,rol,axis=-1)[y1:y2,x1:x2]
z350_wave = np.roll(z350_wave,rol,axis=-1)[y1:y2,x1:x2]

dbz_ctl = smooth2d(dbz_ctl,5)
dbz_adj = smooth2d(dbz_adj,5)
dbz_wave = smooth2d(dbz_wave,5)

dv900_adj = np.nan_to_num(v900_adj - v900_ctl)
dv900_adj = np.nan_to_num(v900_adj - v900_ctl)

dv900_wave = np.nan_to_num(v900_wave - v900_ctl)
dv900_wave = np.nan_to_num(v900_wave - v900_ctl)

dv500_adj = np.nan_to_num(v500_adj - v500_ctl)
dv500_adj = np.nan_to_num(v500_adj - v500_ctl)

dv500_wave = np.nan_to_num(v500_wave - v500_ctl)
dv500_wave = np.nan_to_num(v500_wave - v500_ctl)

dv350_adj = np.nan_to_num(v350_adj - v350_ctl)
dv350_adj = np.nan_to_num(v350_adj - v350_ctl)

dv350_wave = np.nan_to_num(v350_wave - v350_ctl)
dv350_wave = np.nan_to_num(v350_wave - v350_ctl)
```

```python
fig, axd = plt.subplot_mosaic([['top left','top center','top right','cbar'],
                               ['bottom left','bottom center','bottom right','cbar']],
                              constrained_layout=True, figsize=(16.5,9), 
                              gridspec_kw={'width_ratios':[1,1,1,0.1],'height_ratios':[1,1]})

lvls = np.asarray([-9,-7,-5,-3,-1,1,3,5,7,9])
nm = colors.BoundaryNorm(lvls,256)
trans = mtransforms.ScaledTranslation(1/25, -1/25, fig.dpi_scale_trans)

cmap = plt.get_cmap('bwr')
cmap.set_under('navy')
cmap.set_over('brown')

cs1 = axd['top left'].contour(z900_adj,colors='k',levels=np.arange(0,1000,10),linewidths=2,alpha=0.8)
axd['top left'].clabel(cs1,fmt='%1.0f',inline=1,levels=np.arange(0,1000,10),fontsize=10,colors='k')
im1 = axd['top left'].contourf(dv900_adj,levels=lvls,extend='both',cmap=cmap,alpha=0.9,norm=nm)
axd['top left'].contourf(dbz_adj,levels=[0,0.01],colors='k',extend='max',alpha=0.25)
axd['top left'].set_xticks(xax2)
axd['top left'].set_xticklabels(['','','',''])
axd['top left'].set_ylabel('South-North (km)',labelpad=6,fontsize=20)
axd['top left'].set_yticks(yax2)
axd['top left'].set_yticklabels(yax1)
axd['top left'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['top left'].set_aspect('equal')
axd['top left'].set_title('900 hPa',pad=12,fontsize=24)
axd['top left'].grid()
axd['top left'].text(-0.575, 0.625, 'Adjoint',transform=axd['top left'].transAxes + trans,
            fontsize=24, verticalalignment='top',rotation='vertical')
axd['top left'].text(0.001, 1.0, '(a)',transform=axd['top left'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2),zorder=10)

cs2 = axd['top center'].contour(z500_adj,colors='k',levels=np.arange(0,1000,10),linewidths=2,alpha=0.8)
axd['top center'].clabel(cs2,fmt='%1.0f',inline=1,levels=np.arange(0,1000,10),fontsize=10,colors='k')
im2 = axd['top center'].contourf(dv500_adj,levels=lvls,extend='both',cmap=cmap,alpha=0.9,norm=nm)
axd['top center'].contourf(dbz_adj,levels=[0,0.01],colors='k',extend='max',alpha=0.25)
axd['top center'].set_xticks(xax2)
axd['top center'].set_xticklabels(['','','',''])
axd['top center'].set_yticks(yax2)
axd['top center'].set_yticklabels(['','','',''])
axd['top center'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['top center'].set_aspect('equal')
axd['top center'].set_title('500 hPa',pad=12,fontsize=24)
axd['top center'].grid()
axd['top center'].text(0.001, 1.0, '(b)',transform=axd['top center'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2),zorder=10)

cs3 = axd['top right'].contour(z350_adj,colors='k',levels=np.arange(0,1000,10),linewidths=2,alpha=0.8)
axd['top right'].clabel(cs3,fmt='%1.0f',inline=1,levels=np.arange(0,1000,10),fontsize=10,colors='k')
im3 = axd['top right'].contourf(dv350_adj,levels=lvls,extend='both',cmap=cmap,alpha=0.9,norm=nm)
axd['top right'].contourf(dbz_adj,levels=[0,0.01],colors='k',extend='max',alpha=0.25)
axd['top right'].set_xticks(xax2)
axd['top right'].set_xticklabels(['','','',''])
axd['top right'].set_yticks(yax2)
axd['top right'].set_yticklabels(['','','',''])
axd['top right'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['top right'].set_aspect('equal')
axd['top right'].set_title('350 hPa',pad=12,fontsize=24)
axd['top right'].grid()
axd['top right'].text(0.001, 1.0, '(c)',transform=axd['top right'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2),zorder=10)

cs4 = axd['bottom left'].contour(z900_wave,colors='k',levels=np.arange(0,1000,10),linewidths=2,alpha=0.8)
axd['bottom left'].clabel(cs4,fmt='%1.0f',inline=1,levels=np.arange(0,1000,10),fontsize=10,colors='k')
im4 = axd['bottom left'].contourf(dv900_wave,levels=lvls,extend='both',cmap=cmap,alpha=0.9,norm=nm)
axd['bottom left'].contourf(dbz_wave,levels=[0,0.01],colors='k',extend='max',alpha=0.25)
axd['bottom left'].set_xlabel('West-East (km)',labelpad=6,fontsize=20)
axd['bottom left'].set_xticks(xax2)
axd['bottom left'].set_xticklabels(xax1)
axd['bottom left'].set_ylabel('South-North (km)',labelpad=6,fontsize=20)
axd['bottom left'].set_yticks(yax2)
axd['bottom left'].set_yticklabels(yax1)
axd['bottom left'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['bottom left'].set_aspect('equal')
axd['bottom left'].grid()
axd['bottom left'].text(-0.575, 0.625, 'Wave',transform=axd['bottom left'].transAxes + trans,
            fontsize=24, verticalalignment='top',rotation='vertical')
axd['bottom left'].text(0.001, 1.0, '(d)',transform=axd['bottom left'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2),zorder=10)

cs5 = axd['bottom center'].contour(z500_wave,colors='k',levels=np.arange(0,1000,10),linewidths=2,alpha=0.8)
axd['bottom center'].clabel(cs5,fmt='%1.0f',inline=1,levels=np.arange(0,1000,10),fontsize=10,colors='k')
im5 = axd['bottom center'].contourf(dv500_wave,levels=lvls,extend='both',cmap=cmap,alpha=0.9,norm=nm)
axd['bottom center'].contourf(dbz_wave,levels=[0,0.01],colors='k',extend='max',alpha=0.25)
axd['bottom center'].set_xlabel('West-East (km)',labelpad=6,fontsize=20)
axd['bottom center'].set_xticks(xax2)
axd['bottom center'].set_xticklabels(xax1)
axd['bottom center'].set_yticks(yax2)
axd['bottom center'].set_yticklabels(['','','',''])
axd['bottom center'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['bottom center'].set_aspect('equal')
axd['bottom center'].grid()
axd['bottom center'].text(0.001, 1.0, '(e)',transform=axd['bottom center'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2),zorder=10)

cs6 = axd['bottom right'].contour(z350_wave,colors='k',levels=np.arange(0,1000,10),linewidths=2,alpha=0.8)
axd['bottom right'].clabel(cs6,fmt='%1.0f',inline=1,levels=np.arange(0,1000,10),fontsize=10,colors='k')
im6 = axd['bottom right'].contourf(dv350_wave,levels=lvls,extend='both',cmap=cmap,alpha=0.9,norm=nm)
axd['bottom right'].contourf(dbz_wave,levels=[0,0.01],colors='k',extend='max',alpha=0.25)
axd['bottom right'].set_xlabel('West-East (km)',labelpad=6,fontsize=20)
axd['bottom right'].set_xticks(xax2)
axd['bottom right'].set_xticklabels(xax1)
axd['bottom right'].set_yticks(yax2)
axd['bottom right'].set_yticklabels(['','','',''])
axd['bottom right'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['bottom right'].set_aspect('equal')
axd['bottom right'].grid()
axd['bottom right'].text(0.001, 1.0, '(f)',transform=axd['bottom right'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2),zorder=10)

cbar = fig.colorbar(im1, orientation="vertical", cax=axd['cbar'])
cbar.ax.tick_params(labelsize=20, pad=10)

#cbar = fig.colorbar(im4, orientation="vertical", cax=axd['cbar2'])
#cbar.ax.tick_params(labelsize=20, pad=10)

plt.savefig('/p/work1/lloveras/adj_4km/revised_figs/dvlevs_full.pdf',bbox_inches='tight')
```
