```python
%pylab inline
import warnings
warnings.filterwarnings('ignore')
import matplotlib.transforms as mtransforms
from matplotlib import ticker as mtick
import matplotlib.colors as colors
```

```python
t_ctl = np.load('/p/work1/lloveras/adj_4km/proc_short/th_ctl.npy')
t_wave = np.load('/p/work1/lloveras/adj_4km/proc_short/th_wave.npy')

z_ctl = np.load('/p/work1/lloveras/adj_4km/proc_short/z_ctl.npy')
z_wave = np.load('/p/work1/lloveras/adj_4km/proc_short/z_wave.npy')

dbz_ctl = np.load('/p/work1/lloveras/adj_4km/proc_short/dbz_ctl.npy')
dbz_wave = np.load('/p/work1/lloveras/adj_4km/proc_short/dbz_wave.npy')
```

```python
z1 = 0
z2 = 100
x1 = 1375
x2 = 1525
dx = 4

zgrid = np.arange(z1,z2)
ygrid = np.arange(x1*dx,x2*dx,dx)
y,zmesh = np.meshgrid(ygrid,zgrid)
```

```python
fig, axd = plt.subplot_mosaic([['topleft','topright','cbar'],['bottomleft','bottomright','cbar']],
                              constrained_layout=True, figsize=(15,12),dpi=500, 
                              gridspec_kw={'width_ratios':[1,1,0.05],
                                           'height_ratios':[1,1]})

lvls1 = np.asarray([-9,-7,-5,-3,-1,1,3,5,7,9])/5

trans = mtransforms.ScaledTranslation(1/30, -1/30, fig.dpi_scale_trans)

im1 = axd['topleft'].contourf(y,z_wave[0,:,x1:x2]/1000.,(t_wave - t_ctl)[0,:,x1:x2],
                           levels=lvls1,extend='both',cmap=get_cmap('RdBu_r'))
cbar1 = fig.colorbar(im1, orientation='vertical', cax=axd['cbar'])
cbar1.ax.tick_params(labelsize=20, pad=6)
axd['topleft'].contourf(y,z_wave[0,:,x1:x2]/1000.,dbz_ctl[0,:,x1:x2],
                     levels=[0,0.01],colors='k',extend='max',alpha=0.25)
axd['topleft'].tick_params(axis='both', labelsize=20, pad=8,length=8,which='both')
axd['topleft'].set_ylabel('Height (km)',fontsize=20,labelpad=6)
axd['topleft'].yaxis.set_major_locator(MultipleLocator(4))
axd['topleft'].yaxis.set_major_formatter(FormatStrFormatter('%d'))
axd['topleft'].yaxis.set_minor_locator(MultipleLocator(2))
axd['topleft'].set_ylim((0,18))
axd['topleft'].xaxis.set_major_locator(MultipleLocator(200))
axd['topleft'].xaxis.set_major_formatter(FormatStrFormatter('%d'))
axd['topleft'].xaxis.set_minor_locator(MultipleLocator(100))
axd['topleft'].set_xticklabels(['','','','','',''])
axd['topleft'].text(0.0, 1.0, '(a) t = 0',transform=axd['topleft'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='k', pad=2),zorder=50)

axd['topright'].contourf(y,z_wave[1,:,x1:x2]/1000.,(t_wave - t_ctl)[1,:,x1:x2],
                           levels=lvls1,extend='both',cmap=get_cmap('RdBu_r'))
axd['topright'].contourf(y,z_wave[1,:,x1:x2]/1000.,dbz_ctl[1,:,x1:x2],
                     levels=[0,0.01],colors='k',extend='max',alpha=0.25)
axd['topright'].tick_params(axis='both', labelsize=20, pad=8,length=8,which='both')
axd['topright'].yaxis.set_major_locator(MultipleLocator(4))
axd['topright'].yaxis.set_major_formatter(FormatStrFormatter('%d'))
axd['topright'].yaxis.set_minor_locator(MultipleLocator(2))
axd['topright'].set_yticklabels(['','','','','',''])
axd['topright'].set_ylim((0,18))
axd['topright'].xaxis.set_major_locator(MultipleLocator(200))
axd['topright'].xaxis.set_major_formatter(FormatStrFormatter('%d'))
axd['topright'].xaxis.set_minor_locator(MultipleLocator(100))
axd['topright'].set_xticklabels(['','','','','',''])
axd['topright'].text(0.0, 1.0, '(b) t = 0.5 h',transform=axd['topright'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='k', pad=2),zorder=50)

axd['bottomleft'].contourf(y,z_wave[2,:,x1:x2]/1000.,(t_wave - t_ctl)[2,:,x1:x2],
                           levels=lvls1,extend='both',cmap=get_cmap('RdBu_r'))
axd['bottomleft'].contourf(y,z_wave[2,:,x1:x2]/1000.,dbz_ctl[2,:,x1:x2],
                     levels=[0,0.01],colors='k',extend='max',alpha=0.25)
axd['bottomleft'].tick_params(axis='both', labelsize=20, pad=8,length=8,which='both')
axd['bottomleft'].set_ylabel('Height (km)',fontsize=20,labelpad=6)
axd['bottomleft'].set_xlabel('West-East (km)',fontsize=20,labelpad=6)
axd['bottomleft'].yaxis.set_major_locator(MultipleLocator(4))
axd['bottomleft'].yaxis.set_major_formatter(FormatStrFormatter('%d'))
axd['bottomleft'].yaxis.set_minor_locator(MultipleLocator(2))
axd['bottomleft'].set_ylim((0,18))
axd['bottomleft'].xaxis.set_major_locator(MultipleLocator(200))
axd['bottomleft'].xaxis.set_major_formatter(FormatStrFormatter('%d'))
axd['bottomleft'].xaxis.set_minor_locator(MultipleLocator(100))
axd['bottomleft'].text(0.0, 1.0, '(c) t = 1 h',transform=axd['bottomleft'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='k', pad=2),zorder=50)

axd['bottomright'].contourf(y,z_wave[3,:,x1:x2]/1000.,(t_wave - t_ctl)[3,:,x1:x2],
                           levels=lvls1,extend='both',cmap=get_cmap('RdBu_r'))
axd['bottomright'].contourf(y,z_wave[3,:,x1:x2]/1000.,dbz_ctl[3,:,x1:x2],
                     levels=[0,0.01],colors='k',extend='max',alpha=0.25)
axd['bottomright'].tick_params(axis='both', labelsize=20, pad=8,length=8,which='both')
axd['bottomright'].set_xlabel('West-East (km)',fontsize=20,labelpad=6)
axd['bottomright'].yaxis.set_major_locator(MultipleLocator(4))
axd['bottomright'].yaxis.set_major_formatter(FormatStrFormatter('%d'))
axd['bottomright'].yaxis.set_minor_locator(MultipleLocator(2))
axd['bottomright'].set_yticklabels(['','','','','',''])
axd['bottomright'].set_ylim((0,18))
axd['bottomright'].xaxis.set_major_locator(MultipleLocator(200))
axd['bottomright'].xaxis.set_major_formatter(FormatStrFormatter('%d'))
axd['bottomright'].xaxis.set_minor_locator(MultipleLocator(100))
axd['bottomright'].text(0.0, 1.0, '(d) t = 1.5 h',transform=axd['bottomright'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='k', pad=2),zorder=50)

plt.savefig('/p/work1/lloveras/adj_4km/figs_short/wave.png',bbox_inches='tight')
plt.show()
```
