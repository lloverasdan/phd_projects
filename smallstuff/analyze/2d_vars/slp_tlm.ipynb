```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import warnings
warnings.filterwarnings("ignore")
import matplotlib.transforms as mtransforms

G = 9.81
dx = 4
dx_32km = 32
rol = -1000
cent_length = 2500000.

slp_ctl_0h = np.load('/p/work1/lloveras/adj_4km/processed/2d_fields/data_48h/slp_ctl_0h.npy')
slp_ctl_48h = np.load('/p/work1/lloveras/adj_4km/processed/2d_fields/data_48h/slp_ctl_48h.npy')
slp_adj_48h = np.load('/p/work1/lloveras/adj_4km/processed/2d_fields/data_48h/slp_adj_48h.npy')
tlm_48h = np.load('/p/work1/lloveras/adj_4km/tlm_files/tlm_p_4km.npy')

dslp_nlm = np.roll(slp_adj_48h - slp_ctl_48h,rol,axis=-1)
slp_nlm = np.roll(slp_ctl_48h,rol,axis=-1)
dslp_tlm = np.roll(tlm_48h,rol,axis=-1)
```

```python
x1_48 = 750
x2_48 = 1126
y1_48 = 750
y2_48 = 1126

yax1_48 = np.arange(y1_48*dx,y2_48*dx,500).astype(int)
yax2_48 = (yax1_48 - y1_48*dx)/float(dx)
yax2_48 = yax2_48.astype(int)
yax1_48 = ["%0.0f" % number for number in yax1_48]
xax1_48 = np.arange(x1_48*dx,x2_48*dx,500).astype(int)
xax2_48 = (xax1_48 - x1_48*dx)/float(dx)
xax2_48 = xax2_48.astype(int)
xax1_48 = ["%0.0f" % number for number in xax1_48]

slp_nlm_plot = slp_nlm[y1_48:y2_48,x1_48:x2_48]
dslp_nlm_plot = dslp_nlm[y1_48:y2_48,x1_48:x2_48]
dslp_tlm_plot = dslp_tlm[y1_48:y2_48,x1_48:x2_48]
```

```python
fig, axd = plt.subplot_mosaic([['top left','top right'],
                               ['cbar1','cbar2']],
                              constrained_layout=True, figsize=(13.5,7.5), 
                              gridspec_kw={'width_ratios':[1,1],'height_ratios':[1,0.075]})

trans = mtransforms.ScaledTranslation(1/25, -1/25, fig.dpi_scale_trans)

lvls = np.asarray([-9,-7,-5,-3,-1,1,3,5,7,9])*1.5
lvls2 = np.asarray([-9,-7,-5,-3,-1,1,3,5,7,9])*3.5
nm = colors.BoundaryNorm(lvls,256)
nm2 = colors.BoundaryNorm(lvls2,256)

clrs = ['cyan',#'#0000FF',
        '#3434FF',
        '#6666FF',
        '#9898FF',
        '#CCCCFF',
        '#FFFEFE',
        '#FFCCCC',
        '#FF9A9A',
        '#FF6666',
        '#FF3232',
        'yellow'] #'FF0000'
cmap = matplotlib.colors.LinearSegmentedColormap.from_list("",clrs)
cmap.set_under('darkgreen')
cmap.set_over('brown')

cs1 = axd['top left'].contour(slp_nlm_plot,colors='k',levels=np.arange(950,1060,4),linewidths=2,alpha=0.8)
axd['top left'].clabel(cs1,fmt='%1.0f',inline=1,levels=np.arange(950,1060,8),fontsize=10,colors='k')
im1 = axd['top left'].contourf(dslp_nlm_plot,levels=lvls,extend='both',cmap=cmap,norm=nm,alpha=0.9)
axd['top left'].set_xlabel('West-East (km)',labelpad=6,fontsize=20)
axd['top left'].set_xticks(xax2_48)
axd['top left'].set_xticklabels(xax1_48)
axd['top left'].set_ylabel('South-North (km)',labelpad=6,fontsize=20)
axd['top left'].set_yticks(yax2_48)
axd['top left'].set_yticklabels(yax1_48)
axd['top left'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['top left'].set_aspect('equal')
axd['top left'].set_title('Nonlinear',pad=12,fontsize=24)
axd['top left'].grid()
axd['top left'].text(0.001, 1.0, '(a)',transform=axd['top left'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2),zorder=10)

cs3 = axd['top right'].contour(slp_nlm_plot,colors='k',levels=np.arange(950,1060,4),linewidths=2,alpha=0.8)
axd['top right'].clabel(cs3,fmt='%1.0f',inline=1,levels=np.arange(950,1060,8),fontsize=10,colors='k')
im3 = axd['top right'].contourf(dslp_tlm_plot,levels=lvls2,extend='both',cmap=cmap,norm=nm2,alpha=0.9)
axd['top right'].set_xlabel('West-East (km)',labelpad=6,fontsize=20)
axd['top right'].set_xticks(xax2_48)
axd['top right'].set_xticklabels(xax1_48)
axd['top right'].set_yticks(yax2_48)
axd['top right'].set_yticklabels(['','','',''])
axd['top right'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['top right'].set_aspect('equal')
axd['top right'].set_title('Tangent Linear',pad=12,fontsize=24)
axd['top right'].grid()
axd['top right'].text(0.001, 1.0, '(b)',transform=axd['top right'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2),zorder=10)

cbar = fig.colorbar(im1, orientation="horizontal", cax=axd['cbar1'],ticks=lvls)
cbar.ax.tick_params(labelsize=18, pad=10)
cbar.set_ticklabels(['-13.5','','-7.5','','-1.5','1.5','','7.5','','13.5'])

cbar2 = fig.colorbar(im3, orientation="horizontal", cax=axd['cbar2'],ticks=lvls2)
cbar2.ax.tick_params(labelsize=18, pad=10)
cbar2.set_ticklabels(['-31.5','','-17.5','','-3.5','3.5','','17.5','','31.5'])

plt.savefig('/p/work1/lloveras/adj_4km/figs_june/dslp_tlm.pdf',bbox_inches='tight')
```
