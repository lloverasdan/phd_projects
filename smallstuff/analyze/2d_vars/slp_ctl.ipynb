```python
%pylab inline
import matplotlib
import matplotlib.pyplot as plt
from netCDF4 import Dataset
#from wrf import getvar, vinterp, smooth2d
import numpy as np
from matplotlib.cm import get_cmap
from matplotlib.ticker import MultipleLocator
import matplotlib.transforms as mtransforms

nt = 33
slp1 = np.append(np.load('/p/work1/lloveras/adj_4km/processed/2d_fields/slp_ctl1.npy'),np.load('/p/work1/lloveras/adj_4km/processed/2d_fields/slp_ctl2.npy'))
slp1[0] = slp1[1]
slp1[9] = 0.5*(slp1[8] + slp1[10])
slp1[28] = 0.5*(slp1[27] + slp1[29])
slp1[30] = 0.5*(slp1[29] + slp1[31])
tlist=[x*6 for x in list(range(nt))]
dx = 4
rol = -1000

#nc1 = Dataset('/p/work1/lloveras/adj_4km/wrfout_ctl')
```

```python
fig, axd = plt.subplot_mosaic([['top']],
                              constrained_layout=True, figsize=(8,6))

trans = mtransforms.ScaledTranslation(1/25, -1/25, fig.dpi_scale_trans)

axd['top'].plot(tlist,slp1,color='b',linewidth=4.0)
axd['top'].xaxis.set_major_locator(MultipleLocator(24))
axd['top'].xaxis.set_minor_locator(MultipleLocator(12))
axd['top'].tick_params(axis='both',labelsize=20,which='both',direction='in',length=4,width=1)
axd['top'].grid(linewidth=1.5)
axd['top'].set_xlabel('Time (h)',fontsize=20,labelpad=6)
axd['top'].set_ylim(955,1015)
axd['top'].set_yticks([960,970,980,990,1000,1010])
#axd['top'].set_yticklabels(['960','','980','','1000',''])
axd['top'].tick_params(axis='both',labelsize=20,pad=6)
axd['top'].set_ylabel('Minimum Pressure (hPa)',fontsize=20,labelpad=6)
axd['top'].text(0.0, 1.0, '(a)',transform=axd['top'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2))
plt.savefig('/p/work1/lloveras/adj_4km/meso_figs/slp_time.pdf',bbox_inches='tight')
```

```python
slp_48 = np.asarray(getvar(nc1,'slp',timeidx=8))
slp_72 = np.asarray(getvar(nc1,'slp',timeidx=16))
slp_96 = np.asarray(getvar(nc1,'slp',timeidx=24))

t_48 = np.asarray(getvar(nc1,'th',timeidx=8))
t_72 = np.asarray(getvar(nc1,'th',timeidx=16))
t_96 = np.asarray(getvar(nc1,'th',timeidx=24))
```

```python
x1_48 = 0
x2_48 = 1001
y1_48 = 500
y2_48 = 1501

x1_72 = 500
x2_72 = 1501
y1_72 = 500
y2_72 = 1501

x1_96 = 800
x2_96 = 1801
y1_96 = 500
y2_96 = 1501

yax1_48 = np.arange(y1_48*dx,y2_48*dx,1000).astype(int)
yax2_48 = (yax1_48 - y1_48*dx)/float(dx)
yax2_48 = yax2_48.astype(int)
yax1_48 = ["%0.0f" % number for number in yax1_48]
xax1_48 = np.arange(x1_48*dx,x2_48*dx,1000).astype(int)
xax2_48 = (xax1_48 - x1_48*dx)/float(dx)
xax2_48 = xax2_48.astype(int)
xax1_48 = ["%0.0f" % number for number in xax1_48]

yax1_72 = np.arange(y1_72*dx,y2_72*dx,1000).astype(int)
yax2_72 = (yax1_72 - y1_72*dx)/float(dx)
yax2_72 = yax2_72.astype(int)
yax1_72 = ["%0.0f" % number for number in yax1_72]
xax1_72 = np.arange(x1_72*dx,x2_72*dx,1000).astype(int)
xax2_72 = (xax1_72 - x1_72*dx)/float(dx)
xax2_72 = xax2_72.astype(int)
xax1_72 = ["%0.0f" % number for number in xax1_72]

yax1_96 = np.arange(y1_96*dx,y2_96*dx,1000).astype(int)
yax2_96 = (yax1_96 - y1_96*dx)/float(dx)
yax2_96 = yax2_96.astype(int)
yax1_96 = ["%0.0f" % number for number in yax1_96]
xax1_96 = np.arange(x1_96*dx,x2_96*dx,1000).astype(int)
xax2_96 = (xax1_96 - x1_96*dx)/float(dx)
xax2_96 = xax2_96.astype(int)
xax1_96 = ["%0.0f" % number for number in xax1_96]

pplot_48 = np.roll(slp_48,rol,axis=-1)[y1_48:y2_48,x1_48:x2_48]
pplot_72 = np.roll(slp_72,rol,axis=-1)[y1_72:y2_72,x1_72:x2_72]
pplot_96 = np.roll(slp_96,rol,axis=-1)[y1_96:y2_96,x1_96:x2_96]

tplot_48 = np.roll(t_48,rol,axis=-1)[0,y1_48:y2_48,x1_48:x2_48]
tplot_72 = np.roll(t_72,rol,axis=-1)[0,y1_72:y2_72,x1_72:x2_72]
tplot_96 = np.roll(t_96,rol,axis=-1)[0,y1_96:y2_96,x1_96:x2_96]
```

```python
fig, axd = plt.subplot_mosaic([['top', 'top', 'top'],['bottom left', 'bottom center', 'bottom right'],
                               ['cbar','cbar','cbar']],
                              constrained_layout=True, figsize=(16.5,10.0), 
                              gridspec_kw={'height_ratios':[1.5,2,0.15]})

trans = mtransforms.ScaledTranslation(1/25, -1/25, fig.dpi_scale_trans)

axd['top'].plot(tlist,slp1,color='b',linewidth=4.0)
axd['top'].xaxis.set_major_locator(MultipleLocator(24))
axd['top'].xaxis.set_minor_locator(MultipleLocator(12))
axd['top'].tick_params(axis='both',labelsize=20,which='both',direction='in',length=4,width=1)
axd['top'].grid(linewidth=1.5)
axd['top'].set_xlabel('Time (h)',fontsize=20,labelpad=6)
axd['top'].set_ylim(955,1015)
axd['top'].set_yticks([960,970,980,990,1000,1010])
axd['top'].set_yticklabels(['960','','980','','1000',''])
axd['top'].tick_params(axis='both',labelsize=20,pad=6)
axd['top'].set_ylabel('Minimum Pressure (hPa)',fontsize=20,labelpad=6)
axd['top'].text(0.0, 1.0, '(a)',transform=axd['top'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2))

cs48 = axd['bottom left'].contour(pplot_48,colors='k',levels=np.arange(952,1060,4),linewidths=2,alpha=0.8)
axd['bottom left'].clabel(cs48,fmt='%1.0f',inline=1,levels=np.arange(952,1060,8),fontsize=10,colors='k')
im1 = axd['bottom left'].contourf(tplot_48,levels=np.arange(270,292,2),extend='both',cmap=get_cmap('coolwarm'))
axd['bottom left'].set_xlabel('West-East (km)',fontsize=20,labelpad=6)
axd['bottom left'].set_xticks(xax2_48)
axd['bottom left'].set_xticklabels(xax1_48)
axd['bottom left'].set_ylabel('South-North (km)',labelpad=6,fontsize=20)
axd['bottom left'].set_yticks(yax2_48)
axd['bottom left'].set_yticklabels(yax1_48)
axd['bottom left'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['bottom left'].set_aspect('equal')
axd['bottom left'].text(0.0, 1.0, '(b) 48 h',transform=axd['bottom left'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2))

cs72 = axd['bottom center'].contour(pplot_72,colors='k',levels=np.arange(952,1060,4),linewidths=2,alpha=0.8)
axd['bottom center'].clabel(cs72,fmt='%1.0f',inline=1,levels=np.arange(952,1060,8),fontsize=10,colors='k')
im2 = axd['bottom center'].contourf(tplot_72,levels=np.arange(270,292,2),extend='both',cmap=get_cmap('coolwarm'))
axd['bottom center'].set_xlabel('West-East (km)',fontsize=20,labelpad=6)
axd['bottom center'].set_xticks(xax2_72)
axd['bottom center'].set_xticklabels(xax1_72)
axd['bottom center'].set_yticks(yax2_72)
axd['bottom center'].set_yticklabels(['','','','',''])
axd['bottom center'].tick_params(axis='both', which='major', labelsize=20, pad=8,length=8)
axd['bottom center'].set_aspect('equal')
axd['bottom center'].text(0.0, 1.0, '(c) 96 h',transform=axd['bottom center'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2))

cs96 = axd['bottom right'].contour(pplot_96,colors='k',levels=np.arange(952,1060,4),linewidths=2,alpha=0.8)
axd['bottom right'].clabel(cs96,fmt='%1.0f',inline=1,levels=np.arange(952,1060,8),fontsize=10,colors='k')
im3 = axd['bottom right'].contourf(tplot_96,levels=np.arange(270,292,2),extend='both',cmap=get_cmap('coolwarm'))
axd['bottom right'].set_xlabel('West-East (km)',fontsize=20,labelpad=6)
axd['bottom right'].set_xticks(xax2_96)
axd['bottom right'].set_xticklabels(xax1_96)
axd['bottom right'].set_yticks(yax2_96)
axd['bottom right'].set_yticklabels(['','','','',''])
axd['bottom right'].tick_params(axis='both', which='major', labelsize=20, pad=8,length=8)
axd['bottom right'].set_aspect('equal')
axd['bottom right'].text(0.0, 1.0, '(d) 144 h',transform=axd['bottom right'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2))

cbar = fig.colorbar(im1, orientation="horizontal", cax=axd['cbar'])
cbar.ax.tick_params(labelsize=20, pad=6)

plt.savefig('/p/work1/lloveras/adj_4km/figs/fig02_ctlslp.pdf',bbox_inches='tight')
```
