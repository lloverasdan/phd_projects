```python
%pylab inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator
import warnings
warnings.filterwarnings('ignore')
import matplotlib.transforms as mtransforms

nt = 17
time = np.arange(nt)*3

dte_adj = np.load('/p/work1/lloveras/adj_4km/processed/dte/domain_integrated/long_runs/dte_adj_48h.npy')
dte_adj_neg = np.load('/p/work1/lloveras/adj_4km/processed/dte/domain_integrated/long_runs/dte_adj48h_neg.npy')
dte_adj_neg_10 = np.load('/p/work1/lloveras/adj_4km/processed/dte/domain_integrated/long_runs/dte_adj48h_neg_10.npy')
dte_adj_100 = np.load('/p/work1/lloveras/adj_4km/processed/dte/domain_integrated/long_runs/dte_adj_48h_100.npy')
dte_wave = np.load('/p/work1/lloveras/adj_4km/processed/dte/domain_integrated/long_runs/dte_wave_48h.npy')
dte_wave_10 = np.load('/p/work1/lloveras/adj_4km/processed/dte/domain_integrated/long_runs/dte_wave_48h_10.npy')
dte_wave_100 = np.load('/p/work1/lloveras/adj_4km/processed/dte/domain_integrated/long_runs/dte_wave_48h_100.npy')

mag_adj = np.load('/p/work1/lloveras/adj_4km/processed/slp/mag_slp_adj_48h.npy')
mag_adj_neg = np.load('/p/work1/lloveras/adj_4km/processed/slp/mag_slp_adj_48h_neg.npy')
mag_adj_10 = np.load('/p/work1/lloveras/adj_4km/processed/slp/mag_slp_adj_48h_10.npy')
mag_adj_neg_10 = np.load('/p/work1/lloveras/adj_4km/processed/slp/mag_slp_adj_48h_neg_10.npy')
mag_adj_100 = np.load('/p/work1/lloveras/adj_4km/processed/slp/mag_slp_adj_48h_100.npy')
mag_wave = np.load('/p/work1/lloveras/adj_4km/processed/slp/mag_slp_wave_48h.npy')
mag_wave_10 = np.load('/p/work1/lloveras/adj_4km/processed/slp/mag_slp_wave_48h_10.npy')
mag_wave_100 = np.load('/p/work1/lloveras/adj_4km/processed/slp/mag_slp_wave_48h_100.npy')

fss_adj = np.append(1,np.load('/p/work1/lloveras/adj_4km/processed/fss/fss_adj_48h.npy'))
fss_adj_neg = np.append(1,np.load('/p/work1/lloveras/adj_4km/processed/fss/fss_adj_48h_neg.npy'))
fss_adj_10 = np.append(1,np.load('/p/work1/lloveras/adj_4km/processed/fss/fss_adj_48h_10.npy'))
fss_adj_neg_10 = np.append(1,np.load('/p/work1/lloveras/adj_4km/processed/fss/fss_adj_48h_neg_10.npy'))
fss_adj_100 = np.append(1,np.load('/p/work1/lloveras/adj_4km/processed/fss/fss_adj_48h_100.npy'))
fss_wave = np.append(1,np.load('/p/work1/lloveras/adj_4km/processed/fss/fss_wave_48h.npy'))
fss_wave_10 = np.append(1,np.load('/p/work1/lloveras/adj_4km/processed/fss/fss_wave_48h_10.npy'))
fss_wave_100 = np.append(1,np.load('/p/work1/lloveras/adj_4km/processed/fss/fss_wave_48h_100.npy'))

mag_adj[0] = 0.
mag_adj_neg[0] = 0.
mag_adj_10[0] = 0.
mag_adj_neg_10[0] = 0.
mag_adj_100[0] = 0.
mag_wave[0] = 0.
mag_wave_10[0] = 0.
mag_wave_100[0] = 0.

dte_adj_neg[0] = dte_adj[0]
dte_adj_neg_10[0] = dte_adj_10[0]
```

```python
fig, axd = plt.subplot_mosaic([['top left', 'top right'],
                                ['bottom left', 'bottom right']],
                              constrained_layout=True, figsize=(14.0,9.5),dpi=300,
                              gridspec_kw={'width_ratios':[1,1],'height_ratios':[1,1]})

trans = mtransforms.ScaledTranslation(1/10, -1/10, fig.dpi_scale_trans)

axd['top left'].plot(time,dte_adj,color='orange',linewidth=3.0,label='Adjoint')
axd['top left'].plot(time,dte_adj_neg,color='green',linewidth=3.0,label='- Adjoint')
axd['top left'].plot(time,dte_wave,color='blue',linewidth=3.0,label='Wave')
axd['top left'].plot(time,dte_adj_10,color='orange',linewidth=3.0,linestyle='dashed')
axd['top left'].plot(time,dte_adj_neg_10,color='green',linewidth=3.0,linestyle='dashed')
axd['top left'].plot(time,dte_wave_10,color='blue',linewidth=3.0,linestyle='dashed')
axd['top left'].plot(time,dte_adj_100,color='orange',linewidth=3.0,linestyle='dotted')
axd['top left'].plot(time,dte_wave_100,color='blue',linewidth=3.0,linestyle='dotted')
axd['top left'].legend(prop={"size":18})
axd['top left'].set_ylabel('DTE (m$^2$ s$^{-2}$)',labelpad=6,fontsize=20)
axd['top left'].xaxis.set_major_locator(MultipleLocator(6))
axd['top left'].xaxis.set_minor_locator(MultipleLocator(3))
axd['top left'].set_xticklabels(['','','','','','','','','',''])
axd['top left'].tick_params(axis='both',labelsize=20,which='both',direction='in',length=4,width=1)
axd['top left'].legend(loc=4,prop={"size":18})
axd['top left'].set_yscale('log')
axd['top left'].set_ylim([1e0,1e9])
locmaj = matplotlib.ticker.LogLocator(base=10,numticks=10) 
axd['top left'].yaxis.set_major_locator(locmaj)
axd['top left'].set_yticklabels(['','','$10^{1}$','','$10^{3}$','','$10^{5}$','','$10^{7}$','','$10^{9}$'])
locmin = matplotlib.ticker.LogLocator(base=10.0,subs=(0.2,0.4,0.6,0.8),numticks=10)
axd['top left'].yaxis.set_minor_locator(locmin)
axd['top left'].yaxis.set_minor_formatter(matplotlib.ticker.NullFormatter())
axd['top left'].grid(linewidth=1.5)
axd['top left'].text(0.0, 1.0, '(a)',transform=axd['top left'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2))

axd['top right'].plot(time,dte_adj/1e6,color='orange',linewidth=3.0,label='Adjoint')
axd['top right'].plot(time,dte_adj_neg/1e6,color='green',linewidth=3.0,label='- Adjoint')
axd['top right'].plot(time,dte_wave/1e6,color='blue',linewidth=3.0,label='Wave')
axd['top right'].plot(time,dte_adj_10/1e6,color='orange',linewidth=3.0,linestyle='dashed')
axd['top right'].plot(time,dte_adj_neg_10/1e6,color='green',linewidth=3.0,linestyle='dashed')
axd['top right'].plot(time,dte_wave_10/1e6,color='blue',linewidth=3.0,linestyle='dashed')
axd['top right'].plot(time,dte_adj_100/1e6,color='orange',linewidth=3.0,linestyle='dotted')
axd['top right'].plot(time,dte_wave_100/1e6,color='blue',linewidth=3.0,linestyle='dotted')
axd['top right'].set_xticklabels(['','','','','','','','','',''])
axd['top right'].set_ylabel('DTE (10$^6$ m$^2$ s$^{-2}$)',labelpad=6,fontsize=20)
axd['top right'].set_ylim([-1,9])
axd['top right'].xaxis.set_major_locator(MultipleLocator(6))
axd['top right'].xaxis.set_minor_locator(MultipleLocator(3))
axd['top right'].tick_params(axis='both',labelsize=20,which='both',direction='in',length=4,width=1)
axd['top right'].grid(linewidth=1.5)
axd['top right'].text(0.0, 1.0, '(b)',transform=axd['top right'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2))

axd['bottom left'].plot(time,mag_adj,color='orange',label='Adjoint',linewidth=3.0)
axd['bottom left'].plot(time,mag_adj_neg,color='green',label='- Adjoint',linewidth=3.0)
axd['bottom left'].plot(time,mag_adj_10,color='orange',linestyle='dashed',linewidth=3.0)
axd['bottom left'].plot(time,mag_adj_neg_10,color='green',linestyle='dashed',linewidth=3.0)
axd['bottom left'].plot(time,mag_adj_100,color='orange',linestyle='dotted',linewidth=3.0)
axd['bottom left'].plot(time,mag_wave,color='blue',label='Wave',linewidth=3.0)
axd['bottom left'].plot(time,mag_wave_10,color='blue',linestyle='dashed',linewidth=3.0)
axd['bottom left'].plot(time,mag_wave_100,color='blue',linestyle='dotted',linewidth=3.0)
axd['bottom left'].set_xlabel('Lead Time (h)',labelpad=6,fontsize=20)
axd['bottom left'].set_ylabel('Pressure Difference (hPa)',labelpad=6,fontsize=20)
axd['bottom left'].xaxis.set_major_locator(MultipleLocator(6))
axd['bottom left'].xaxis.set_minor_locator(MultipleLocator(3))
axd['bottom left'].tick_params(axis='both',labelsize=20,which='both',direction='in',length=4,width=1)
axd['bottom left'].grid(linewidth=1.5)
axd['bottom left'].set_ylim([-9.5,6.5])
axd['bottom left'].text(0.0, 1.0, '(c)',transform=axd['bottom left'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2))

axd['bottom right'].plot(time,fss_adj,color='orange',label='Adjoint',linewidth=3.0)
axd['bottom right'].plot(time,fss_adj_neg,color='green',label='- Adjoint',linewidth=3.0)
axd['bottom right'].plot(time,fss_adj_10,color='orange',linestyle='dashed',linewidth=3.0)
axd['bottom right'].plot(time,fss_adj_neg_10,color='green',linestyle='dashed',linewidth=3.0)
axd['bottom right'].plot(time,fss_adj_100,color='orange',linestyle='dotted',linewidth=3.0)
axd['bottom right'].plot(time,fss_wave,color='blue',label='Wave',linewidth=3.0)
axd['bottom right'].plot(time,fss_wave_10,color='blue',linestyle='dashed',linewidth=3.0)
axd['bottom right'].plot(time,fss_wave_100,color='blue',linestyle='dotted',linewidth=3.0)
axd['bottom right'].set_xlabel('Lead Time (h)',labelpad=6,fontsize=20)
axd['bottom right'].set_ylabel('FSS',labelpad=6,fontsize=20)
axd['bottom right'].xaxis.set_major_locator(MultipleLocator(6))
axd['bottom right'].xaxis.set_minor_locator(MultipleLocator(3))
axd['bottom right'].tick_params(axis='both',labelsize=20,which='both',direction='in',length=4,width=1)
axd['bottom right'].grid(linewidth=1.5)
axd['bottom right'].set_ylim([-0.12,1.12])
axd['bottom right'].text(0.0, 1.0, '(d)',transform=axd['bottom right'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='none', pad=2))

plt.savefig('/p/work1/lloveras/adj_4km/figs/new_figs/norms_negative.pdf',bbox_inches='tight')
```
