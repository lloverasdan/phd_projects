```python
[dudy900_adj, dudx900_adj] = np.gradient(np.nan_to_num(u900_adj - u900_ctl),dx*1000,dx*1000)
[dvdy900_adj, dvdx900_adj] = np.gradient(np.nan_to_num(v900_adj - v900_ctl),dx*1000,dx*1000)

ddiv900_adj = dudx900_adj + dvdy900_adj
dvort900_adj = dvdx900_adj - dudy900_adj

[dudy900_wave, dudx900_wave] = np.gradient(np.nan_to_num(u900_wave - u900_ctl),dx*1000,dx*1000)
[dvdy900_wave, dvdx900_wave] = np.gradient(np.nan_to_num(v900_wave - v900_ctl),dx*1000,dx*1000)

ddiv900_wave = dudx900_wave + dvdy900_wave
dvort900_wave = dvdx900_wave - dudy900_wave

[dudy500_adj, dudx500_adj] = np.gradient(np.nan_to_num(u500_adj - u500_ctl),dx*1000,dx*1000)
[dvdy500_adj, dvdx500_adj] = np.gradient(np.nan_to_num(v500_adj - v500_ctl),dx*1000,dx*1000)

ddiv500_adj = dudx500_adj + dvdy500_adj
dvort500_adj = dvdx500_adj - dudy500_adj

[dudy500_wave, dudx500_wave] = np.gradient(np.nan_to_num(u500_wave - u500_ctl),dx*1000,dx*1000)
[dvdy500_wave, dvdx500_wave] = np.gradient(np.nan_to_num(v500_wave - v500_ctl),dx*1000,dx*1000)

ddiv500_wave = dudx500_wave + dvdy500_wave
dvort500_wave = dvdx500_wave - dudy500_wave

[dudy350_adj, dudx350_adj] = np.gradient(np.nan_to_num(u350_adj - u350_ctl),dx*1000,dx*1000)
[dvdy350_adj, dvdx350_adj] = np.gradient(np.nan_to_num(v350_adj - v350_ctl),dx*1000,dx*1000)

ddiv350_adj = dudx350_adj + dvdy350_adj
dvort350_adj = dvdx350_adj - dudy350_adj

[dudy350_wave, dudx350_wave] = np.gradient(np.nan_to_num(u350_wave - u350_ctl),dx*1000,dx*1000)
[dvdy350_wave, dvdx350_wave] = np.gradient(np.nan_to_num(v350_wave - v350_ctl),dx*1000,dx*1000)

ddiv350_wave = dudx350_wave + dvdy350_wave
dvort350_wave = dvdx350_wave - dudy350_wave
```

```python
### Spectral filter
kx = np.fft.fftfreq(nx,dx)
ky = np.fft.fftfreq(ny,dx)
peak = 1/lam

ddiv900_adj_fft = np.fft.fft2(ddiv900_adj)
ddiv900_adj_fft[np.abs(ky) > peak, :] = 0.
ddiv900_adj_fft[:,np.abs(kx) > peak] = 0.
ddiv900_adj_filt = np.real(np.fft.ifft2(ddiv900_adj_fft))

dvort900_adj_fft = np.fft.fft2(dvort900_adj)
dvort900_adj_fft[np.abs(ky) > peak, :] = 0.
dvort900_adj_fft[:,np.abs(kx) > peak] = 0.
dvort900_adj_filt = np.real(np.fft.ifft2(dvort900_adj_fft))

ddiv900_wave_fft = np.fft.fft2(ddiv900_wave)
ddiv900_wave_fft[np.abs(ky) > peak, :] = 0.
ddiv900_wave_fft[:,np.abs(kx) > peak] = 0.
ddiv900_wave_filt = np.real(np.fft.ifft2(ddiv900_wave_fft))

dvort900_wave_fft = np.fft.fft2(dvort900_wave)
dvort900_wave_fft[np.abs(ky) > peak, :] = 0.
dvort900_wave_fft[:,np.abs(kx) > peak] = 0.
dvort900_wave_filt = np.real(np.fft.ifft2(dvort900_wave_fft))

ddiv500_adj_fft = np.fft.fft2(ddiv500_adj)
ddiv500_adj_fft[np.abs(ky) > peak, :] = 0.
ddiv500_adj_fft[:,np.abs(kx) > peak] = 0.
ddiv500_adj_filt = np.real(np.fft.ifft2(ddiv500_adj_fft))

dvort500_adj_fft = np.fft.fft2(dvort500_adj)
dvort500_adj_fft[np.abs(ky) > peak, :] = 0.
dvort500_adj_fft[:,np.abs(kx) > peak] = 0.
dvort500_adj_filt = np.real(np.fft.ifft2(dvort500_adj_fft))

ddiv500_wave_fft = np.fft.fft2(ddiv500_wave)
ddiv500_wave_fft[np.abs(ky) > peak, :] = 0.
ddiv500_wave_fft[:,np.abs(kx) > peak] = 0.
ddiv500_wave_filt = np.real(np.fft.ifft2(ddiv500_wave_fft))

dvort500_wave_fft = np.fft.fft2(dvort500_wave)
dvort500_wave_fft[np.abs(ky) > peak, :] = 0.
dvort500_wave_fft[:,np.abs(kx) > peak] = 0.
dvort500_wave_filt = np.real(np.fft.ifft2(dvort500_wave_fft))

ddiv350_adj_fft = np.fft.fft2(ddiv350_adj)
ddiv350_adj_fft[np.abs(ky) > peak, :] = 0.
ddiv350_adj_fft[:,np.abs(kx) > peak] = 0.
ddiv350_adj_filt = np.real(np.fft.ifft2(ddiv350_adj_fft))

dvort350_adj_fft = np.fft.fft2(dvort350_adj)
dvort350_adj_fft[np.abs(ky) > peak, :] = 0.
dvort350_adj_fft[:,np.abs(kx) > peak] = 0.
dvort350_adj_filt = np.real(np.fft.ifft2(dvort350_adj_fft))

ddiv350_wave_fft = np.fft.fft2(ddiv350_wave)
ddiv350_wave_fft[np.abs(ky) > peak, :] = 0.
ddiv350_wave_fft[:,np.abs(kx) > peak] = 0.
ddiv350_wave_filt = np.real(np.fft.ifft2(ddiv350_wave_fft))

dvort350_wave_fft = np.fft.fft2(dvort350_wave)
dvort350_wave_fft[np.abs(ky) > peak, :] = 0.
dvort350_wave_fft[:,np.abs(kx) > peak] = 0.
dvort350_wave_filt = np.real(np.fft.ifft2(dvort350_wave_fft))
```
