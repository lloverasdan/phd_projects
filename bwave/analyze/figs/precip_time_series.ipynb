```python
import warnings
warnings.filterwarnings('ignore')
import matplotlib.transforms as mtransforms
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

nt = 33
times = np.arange(nt)*6/24

avg_standard_20km = np.append(0,np.load('/p/work1/lloveras/bwave/processed/precip_avg/standard_20km.npy'))
avg_surface_20km = np.append(0,np.load('/p/work1/lloveras/bwave/processed/precip_avg/surface_20km.npy'))
avg_barotropic_20km = np.append(0,np.load('/p/work1/lloveras/bwave/processed/precip_avg/barotropic_20km.npy'))
avg_standard_4km = np.append(0,np.load('/p/work1/lloveras/bwave/processed/precip_avg/standard_4km.npy'))
avg_surface_4km = np.append(0,np.load('/p/work1/lloveras/bwave/processed/precip_avg/surface_4km.npy'))
avg_barotropic_4km = np.append(0,np.load('/p/work1/lloveras/bwave/processed/precip_avg/barotropic_4km.npy'))

area_standard_20km = np.append(0,np.load('/p/work1/lloveras/bwave/processed/precip_area/standard_20km.npy'))*400
area_surface_20km = np.append(0,np.load('/p/work1/lloveras/bwave/processed/precip_area/surface_20km.npy'))*400
area_barotropic_20km = np.append(0,np.load('/p/work1/lloveras/bwave/processed/precip_area/barotropic_20km.npy'))*400
area_standard_4km = np.append(0,np.load('/p/work1/lloveras/bwave/processed/precip_area/standard_4km.npy'))*16
area_surface_4km = np.append(0,np.load('/p/work1/lloveras/bwave/processed/precip_area/surface_4km.npy'))*16
area_barotropic_4km = np.append(0,np.load('/p/work1/lloveras/bwave/processed/precip_area/barotropic_4km.npy'))*16
```

```python
fig, axd = plt.subplot_mosaic([['left','right']],
                              constrained_layout=True, figsize=(15,7))

trans = mtransforms.ScaledTranslation(1/30, -1/30, fig.dpi_scale_trans)

axd['left'].plot(times,avg_standard_20km,color='blue',linewidth=3.2,linestyle='dashed',label='Neutral, 20 km')
axd['left'].plot(times,avg_barotropic_20km,color='green',linewidth=3.2,linestyle='dashed',label='Barotropic, 20 km')
axd['left'].plot(times,avg_surface_20km,color='orange',linewidth=3.2,linestyle='dashed',label='Low-level, 20 km')

axd['left'].plot(times,avg_standard_4km,color='blue',linewidth=3.2,label='Neutral, 4 km')
axd['left'].plot(times,avg_barotropic_4km,color='green',linewidth=3.2,label='Barotropic, 4 km')
axd['left'].plot(times,avg_surface_4km,color='orange',linewidth=3.2,label='Low-level, 4km')

axd['left'].legend(fontsize=16,loc=1,ncol=2,columnspacing=0.5)
axd['left'].grid()
axd['left'].set_xlabel('Day',fontsize=20,labelpad=6)
axd['left'].xaxis.set_major_locator(MultipleLocator(1))
axd['left'].xaxis.set_minor_locator(MultipleLocator(0.5))
axd['left'].yaxis.set_major_locator(MultipleLocator(0.1))
axd['left'].yaxis.set_minor_locator(MultipleLocator(0.05))
axd['left'].set_ylim([-0.045,0.645])
axd['left'].set_ylabel('Precipitation Intensity (mm h$^{-1}$)',fontsize=20,labelpad=6)
axd['left'].tick_params(axis='both', labelsize=20, which='both', pad=8,length=8)
axd['left'].text(0.0, 1.0, '(a)',transform=axd['left'].transAxes + trans,
            fontsize=24, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='k', pad=2))

axd['right'].plot(times,area_standard_4km/1e6,color='blue',linewidth=3.2)
axd['right'].plot(times,area_barotropic_4km/1e6,color='green',linewidth=3.2)
axd['right'].plot(times,area_surface_4km/1e6,color='orange',linewidth=3.2)
axd['right'].plot(times,area_standard_20km/1e6,color='blue',linewidth=3.2,linestyle='dashed')
axd['right'].plot(times,area_barotropic_20km/1e6,color='green',linewidth=3.2,linestyle='dashed')
axd['right'].plot(times,area_surface_20km/1e6,color='orange',linewidth=3.2,linestyle='dashed')
axd['right'].grid()
axd['right'].set_xlabel('Day',fontsize=20,labelpad=6)
axd['right'].xaxis.set_major_locator(MultipleLocator(1))
axd['right'].xaxis.set_minor_locator(MultipleLocator(0.5))
axd['right'].yaxis.set_major_locator(MultipleLocator(4))
axd['right'].yaxis.set_minor_locator(MultipleLocator(2))
axd['right'].set_ylim([-1.5,17.5])
axd['right'].set_ylabel('Area of Precipitation ($10^6$ km$^2$)',fontsize=20,labelpad=6)
axd['right'].tick_params(axis='both', labelsize=20, which='both', pad=8,length=8)
axd['right'].text(0.0, 1.0, '(b)',transform=axd['right'].transAxes + trans,
            fontsize=24, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='k', pad=2))

plt.savefig('/p/work1/lloveras/bwave/new_figs/precip_time_series.pdf',bbox_inches='tight')
plt.show()
```
