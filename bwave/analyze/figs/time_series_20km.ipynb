```python
import warnings
warnings.filterwarnings('ignore')
import matplotlib.transforms as mtransforms
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

nt = 33
times = np.arange(nt)*6/24

std_dry = np.load('/p/work1/lloveras/bwave/processed/slp_time/standard_dry_20km.npy')
std_dry[1:] = std_dry[1:]+1.5
std_85 = np.load('/p/work1/lloveras/bwave/processed/slp_time/standard_moist_20km.npy')
std_75 = np.load('/p/work1/lloveras/bwave/processed/slp_time/standard_moist_rh75_20km.npy')
std_65 = np.load('/p/work1/lloveras/bwave/processed/slp_time/standard_moist_rh65_20km.npy')

sur_dry = np.load('/p/work1/lloveras/bwave/processed/slp_time/surface_dry_20km.npy')
sur_dry[1:] = sur_dry[1:]+1.5
sur_85 = np.load('/p/work1/lloveras/bwave/processed/slp_time/surface_moist_20km.npy')
sur_75 = np.load('/p/work1/lloveras/bwave/processed/slp_time/surface_moist_rh75_20km.npy')
sur_65 = np.load('/p/work1/lloveras/bwave/processed/slp_time/surface_moist_rh65_20km.npy')

bar_dry = np.load('/p/work1/lloveras/bwave/processed/slp_time/barotropic_dry_20km.npy')
bar_dry[1:] = bar_dry[1:]+1.5
bar_85 = np.load('/p/work1/lloveras/bwave/processed/slp_time/barotropic_moist_20km.npy')
bar_75 = np.load('/p/work1/lloveras/bwave/processed/slp_time/barotropic_moist_rh75_20km.npy')
bar_65 = np.load('/p/work1/lloveras/bwave/processed/slp_time/barotropic_moist_rh65_20km.npy')
```

```python
fig, axd = plt.subplot_mosaic([['left','middle','right']],
                              constrained_layout=True, figsize=(14.5,6),dpi=500)

trans = mtransforms.ScaledTranslation(1/30, -1/30, fig.dpi_scale_trans)

# Neutral
axd['left'].plot(times,std_dry,color='black',linewidth=3.2,label='Dry')
axd['left'].plot(times,std_65,color='green',linewidth=3.2,label='65%')
axd['left'].plot(times,std_75,color='orange',linewidth=3.2,label='75%')
axd['left'].plot(times,std_85,color='blue',linewidth=3.2,label='85%')
axd['left'].legend(fontsize=16,loc=3,ncol=2,columnspacing=0.75)
axd['left'].grid()
axd['left'].set_ylim([-68,4])
axd['left'].set_xlim([-0.25,8.25])
axd['left'].set_xlabel('Day',fontsize=20,labelpad=6)
axd['left'].xaxis.set_major_locator(MultipleLocator(1))
axd['left'].xaxis.set_minor_locator(MultipleLocator(0.5))
axd['left'].yaxis.set_major_locator(MultipleLocator(10))
axd['left'].yaxis.set_minor_locator(MultipleLocator(5))
axd['left'].set_ylabel('$(\Delta \mathrm{SLP})_{\mathrm{min}}$ (hPa)',fontsize=20,labelpad=6)
axd['left'].tick_params(axis='both', labelsize=20, which='both', pad=8,length=8)
axd['left'].text(0.0, 1.0, '(a) Neutral',transform=axd['left'].transAxes + trans,
            fontsize=18, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='k', pad=2))

# Barotropic
axd['middle'].plot(times,bar_dry,color='black',linewidth=3.2,label='Dry')
axd['middle'].plot(times,bar_65,color='green',linewidth=3.2,label='65%')
axd['middle'].plot(times,bar_75,color='orange',linewidth=3.2,label='75%')
axd['middle'].plot(times,bar_85,color='blue',linewidth=3.2,label='85%')
axd['middle'].grid()
axd['middle'].set_ylim([-68,4])
axd['middle'].set_xlim([-0.25,8.25])
axd['middle'].set_xlabel('Day',fontsize=20,labelpad=6)
axd['middle'].xaxis.set_major_locator(MultipleLocator(1))
axd['middle'].xaxis.set_minor_locator(MultipleLocator(0.5))
axd['middle'].yaxis.set_major_locator(MultipleLocator(10))
axd['middle'].yaxis.set_minor_locator(MultipleLocator(5))
axd['middle'].set_yticklabels(['','','','','','','',''])
axd['middle'].tick_params(axis='both', labelsize=20, which='both', pad=8,length=8)
axd['middle'].text(0.0, 1.0, '(b) Barotropic',transform=axd['middle'].transAxes + trans,
            fontsize=18, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='k', pad=2))

# Surface
axd['right'].plot(times,sur_dry,color='black',linewidth=3.2,label='Dry')
axd['right'].plot(times,sur_65,color='green',linewidth=3.2,label='65%')
axd['right'].plot(times,sur_75,color='orange',linewidth=3.2,label='75%')
axd['right'].plot(times,sur_85,color='blue',linewidth=3.2,label='85%')
axd['right'].grid()
axd['right'].set_ylim([-68,4])
axd['right'].set_xlim([-0.25,8.25])
axd['right'].set_xlabel('Day',fontsize=20,labelpad=6)
axd['right'].xaxis.set_major_locator(MultipleLocator(1))
axd['right'].xaxis.set_minor_locator(MultipleLocator(0.5))
axd['right'].yaxis.set_major_locator(MultipleLocator(10))
axd['right'].yaxis.set_minor_locator(MultipleLocator(5))
axd['right'].set_yticklabels(['','','','','','','',''])
axd['right'].tick_params(axis='both', labelsize=20, which='both', pad=8,length=8)
axd['right'].text(0.0, 1.0, '(c) Low-level',transform=axd['right'].transAxes + trans,
            fontsize=18, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='k', pad=2))

plt.savefig('/p/work1/lloveras/bwave/new_figs/slp_time_series_rh.pdf',bbox_inches='tight')
plt.show()
```
