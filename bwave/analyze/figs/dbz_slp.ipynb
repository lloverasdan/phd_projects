```python
%pylab inline
import warnings
warnings.filterwarnings('ignore')
import matplotlib.transforms as mtransforms

slp_all = np.load('/p/work1/lloveras/bwave_nov/processed/2d_vars/slp_surface_4km.npy')
dbz_all = np.load('/p/work1/lloveras/bwave_nov/processed/2d_vars/dbz_surface_4km.npy')
```

```python
### Grab time stamp
slp_day2 = slp_all[3]
dbz_day2 = dbz_all[3]

slp_day3 = slp_all[4]
dbz_day3 = dbz_all[4]

slp_day4 = slp_all[5]
dbz_day4 = dbz_all[5]

slp_day5 = slp_all[6]
dbz_day5 = dbz_all[6]

### Parameters in km
rol1_km = 3500
rol2_km = 3500
x1_km = 0
x2_km = 5000
y1_km = 1000
y2_km = 6000
x_int_km = 1000
y_int_km = 1000

### Contour levels
plevs = np.arange(944,1060,8)
plabels = np.arange(944,1060,8)
dbzlevs = np.arange(0,50,5)

### Axis indices
dx = 4
x1 = int(x1_km/dx)
x2 = int(x2_km/dx)+1
y1 = int(y1_km/dx)
y2 = int(y2_km/dx)+1
rol1 = int(rol1_km/dx)
rol2 = int(rol2_km/dx)

### Add ticks and labels
yax1 = np.arange(y1*dx,y2*dx,y_int_km).astype(int)
yax2 = (yax1 - y1*dx)/float(dx)
yax2 = yax2.astype(int)
xax1 = np.arange(x1*dx,x2*dx,x_int_km).astype(int)
xax2 = (xax1 - x1*dx)/float(dx)
xax2 = xax2.astype(int)

### Adjust the arrays
slp_day2_rol = np.roll(slp_day2,rol1,axis=1)[y1:y2,x1:x2]
dbz_day2_rol = np.roll(dbz_day2,rol1,axis=1)[y1:y2,x1:x2]
slp_day3_rol = np.roll(slp_day3,rol1,axis=1)[y1:y2,x1:x2]
dbz_day3_rol = np.roll(dbz_day3,rol1,axis=1)[y1:y2,x1:x2]
slp_day4_rol = np.roll(slp_day4,rol2,axis=1)[y1:y2,x1:x2]
dbz_day4_rol = np.roll(dbz_day4,rol2,axis=1)[y1:y2,x1:x2]
slp_day5_rol = np.roll(slp_day5,rol2,axis=1)[y1:y2,x1:x2]
dbz_day5_rol = np.roll(dbz_day5,rol2,axis=1)[y1:y2,x1:x2]

### Make the plot
fig, axd = plt.subplot_mosaic([['top left','top right'],['bottom left','bottom right'],['cbar','cbar']],
                              constrained_layout=True, figsize=(12,12), 
                              gridspec_kw={'height_ratios':[1,1,0.05]},dpi=400)

trans = mtransforms.ScaledTranslation(1/25, -1/25, fig.dpi_scale_trans)

cs1 = axd['top left'].contour(slp_day2_rol,colors='k',levels=plevs,linewidths=2,alpha=0.8)
axd['top left'].clabel(cs1,fmt='%1.0f',inline=1,levels=plabels,fontsize=10,colors='k')
im1 = axd['top left'].contourf(dbz_day2_rol,levels=dbzlevs,extend='max',cmap=get_cmap('turbo'))
axd['top left'].set_xticks(xax2)
axd['top left'].set_xticklabels(['','','','','',''])
axd['top left'].set_ylabel('South-North (km)',labelpad=6,fontsize=20)
axd['top left'].set_yticks(yax2)
axd['top left'].set_yticklabels(yax1)
axd['top left'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['top left'].set_aspect('equal')
axd['top left'].grid()
axd['top left'].text(0.0, 1.0, '(a) Day 3',transform=axd['top left'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='k', pad=2))

cs2 = axd['top right'].contour(slp_day3_rol,colors='k',levels=plevs,linewidths=2,alpha=0.8)
axd['top right'].clabel(cs2,fmt='%1.0f',inline=1,levels=plabels,fontsize=10,colors='k')
im2 = axd['top right'].contourf(dbz_day3_rol,levels=dbzlevs,extend='max',cmap=get_cmap('turbo'))
axd['top right'].set_xticks(xax2)
axd['top right'].set_xticklabels(['','','','','',''])
axd['top right'].set_yticks(yax2)
axd['top right'].set_yticklabels(['','','','','',''])
axd['top right'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['top right'].set_aspect('equal')
axd['top right'].grid()
axd['top right'].text(0.0, 1.0, '(b) Day 4',transform=axd['top right'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='k', pad=2))

cs1 = axd['bottom left'].contour(slp_day4_rol,colors='k',levels=plevs,linewidths=2,alpha=0.8)
axd['bottom left'].clabel(cs1,fmt='%1.0f',inline=1,levels=plabels,fontsize=10,colors='k')
im1 = axd['bottom left'].contourf(dbz_day4_rol,levels=dbzlevs,extend='max',cmap=get_cmap('turbo'))
axd['bottom left'].set_xlabel('West-East (km)',fontsize=20,labelpad=6)
axd['bottom left'].set_xticks(xax2)
axd['bottom left'].set_xticklabels(xax1)
axd['bottom left'].set_ylabel('South-North (km)',labelpad=6,fontsize=20)
axd['bottom left'].set_yticks(yax2)
axd['bottom left'].set_yticklabels(yax1)
axd['bottom left'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['bottom left'].set_aspect('equal')
axd['bottom left'].grid()
axd['bottom left'].text(0.0, 0.945, '(c) Day 5',transform=axd['bottom left'].transAxes + trans,
            fontsize=20, verticalalignment='bottom',
            bbox=dict(facecolor='white', edgecolor='k', pad=2))

cs2 = axd['bottom right'].contour(slp_day5_rol,colors='k',levels=plevs,linewidths=2,alpha=0.8)
axd['bottom right'].clabel(cs2,fmt='%1.0f',inline=1,levels=plabels,fontsize=10,colors='k')
im2 = axd['bottom right'].contourf(dbz_day5_rol,levels=dbzlevs,extend='max',cmap=get_cmap('turbo'))
axd['bottom right'].set_xlabel('West-East (km)',fontsize=20,labelpad=6)
axd['bottom right'].set_xticks(xax2)
axd['bottom right'].set_xticklabels(xax1)
axd['bottom right'].set_yticks(yax2)
axd['bottom right'].set_yticklabels(['','','','','',''])
axd['bottom right'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['bottom right'].set_aspect('equal')
axd['bottom right'].grid()
axd['bottom right'].text(0.0, 0.945, '(d) Day 6',transform=axd['bottom right'].transAxes + trans,
            fontsize=20, verticalalignment='bottom',
            bbox=dict(facecolor='white', edgecolor='k', pad=2))

cbar = fig.colorbar(im2, orientation="horizontal", cax=axd['cbar'])
cbar.ax.tick_params(labelsize=20, pad=6)

#plt.savefig('/p/work1/lloveras/bwave_nov/figs_paper/dbz_lowlevel.pdf',bbox_inches='tight')
plt.show()
```
