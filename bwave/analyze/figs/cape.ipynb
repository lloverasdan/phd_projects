```python
%pylab inline
import warnings
warnings.filterwarnings('ignore')
import matplotlib.transforms as mtransforms

slp_standard = np.load('/p/work1/lloveras/bwave_nov/processed/2d_vars/slp_standard_4km.npy')[5]
cape_standard = np.load('/p/work1/lloveras/bwave_nov/processed/2d_vars/cape_standard_4km.npy')[5]

slp_barotropic = np.load('/p/work1/lloveras/bwave_nov/processed/2d_vars/slp_barotropic_4km.npy')[5]
cape_barotropic = np.load('/p/work1/lloveras/bwave_nov/processed/2d_vars/cape_barotropic_4km.npy')[5]

slp_surface = np.load('/p/work1/lloveras/bwave_nov/processed/2d_vars/slp_surface_4km.npy')[5]
cape_surface = np.load('/p/work1/lloveras/bwave_nov/processed/2d_vars/cape_surface_4km.npy')[5]
```

```python
### Parameters in km
x1_km = 0
x2_km = 5000
y1_km = 0
y2_km = 5000
x_int_km = 1000
y_int_km = 1000

### Contour levels
plevs = np.arange(944,1060,8)
plabels = np.arange(944,1060,8)
clevs = np.arange(0,500,50)

### Axis indices
dx = 4
x1 = int(x1_km/dx)
x2 = int(x2_km/dx)+1
y1 = int(y1_km/dx)
y2 = int(y2_km/dx)+1

### Add ticks and labels
yax1 = np.arange(y1*dx,y2*dx,y_int_km).astype(int)
yax2 = (yax1 - y1*dx)/float(dx)
yax2 = yax2.astype(int)
xax1 = np.arange(x1*dx,x2*dx,x_int_km).astype(int)
xax2 = (xax1 - x1*dx)/float(dx)
xax2 = xax2.astype(int)

### Roll
rol1_km = 2000
rol1 = int(rol1_km/dx)
slp_std_rol = np.roll(slp_standard,rol1,axis=1)[y1:y2,x1:x2]
cape_std_rol = np.roll(cape_standard,rol1,axis=1)[y1:y2,x1:x2]

rol2_km = 3000
rol2 = int(rol2_km/dx)
slp_bar_rol = np.roll(slp_barotropic,rol2,axis=1)[y1:y2,x1:x2]
cape_bar_rol = np.roll(cape_barotropic,rol2,axis=1)[y1:y2,x1:x2]

rol3_km = 3000
rol3 = int(rol3_km/dx)
slp_sfc_rol = np.roll(slp_surface,rol3,axis=1)[y1:y2,x1:x2]
cape_sfc_rol = np.roll(cape_surface,rol3,axis=1)[y1:y2,x1:x2]
```

```python
fig, axd = plt.subplot_mosaic([['left','middle','right'],['cbar','cbar','cbar']],
                              constrained_layout=True, figsize=(16,6.5), 
                              gridspec_kw={'width_ratios':[1,1,1],
                                           'height_ratios':[1,0.06]})

trans = mtransforms.ScaledTranslation(1/30, -1/30, fig.dpi_scale_trans)

### Standard CAPE

cs1 = axd['left'].contour(slp_std_rol,colors='k',levels=plevs,linewidths=2,alpha=0.8)
axd['left'].clabel(cs1,fmt='%1.0f',inline=1,levels=plabels,fontsize=10,colors='k')
im1 = axd['left'].contourf(cape_std_rol,levels=clevs,extend='max',cmap=get_cmap('turbo'))
axd['left'].set_xlabel('West-East (km)',fontsize=20,labelpad=6)
axd['left'].set_ylabel('South-North (km)',fontsize=20,labelpad=6)
axd['left'].set_xticks(xax2)
axd['left'].set_xticklabels(xax1)
axd['left'].set_yticks(yax2)
axd['left'].set_yticklabels(yax1)
axd['left'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['left'].set_aspect('equal')
axd['left'].grid()
axd['left'].text(0.0, 1.0, '(a) Neutral',transform=axd['left'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='k', pad=2))

### Barotropic CAPE

cs1 = axd['middle'].contour(slp_bar_rol,colors='k',levels=plevs,linewidths=2,alpha=0.8)
axd['middle'].clabel(cs1,fmt='%1.0f',inline=1,levels=plabels,fontsize=10,colors='k')
im1 = axd['middle'].contourf(cape_bar_rol,levels=clevs,extend='max',cmap=get_cmap('turbo'))
axd['middle'].set_xlabel('West-East (km)',fontsize=20,labelpad=6)
axd['middle'].set_xticks(xax2)
axd['middle'].set_xticklabels(xax1)
axd['middle'].set_yticks(yax2)
axd['middle'].set_yticklabels(['','','','','',''])
axd['middle'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['middle'].set_aspect('equal')
axd['middle'].grid()
axd['middle'].text(0.0, 1.0, '(b) Barotropic',transform=axd['middle'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='k', pad=2))

### Surface CAPE

cs1 = axd['right'].contour(slp_sfc_rol,colors='k',levels=plevs,linewidths=2,alpha=0.8)
axd['right'].clabel(cs1,fmt='%1.0f',inline=1,levels=plabels,fontsize=10,colors='k')
im1 = axd['right'].contourf(cape_sfc_rol,levels=clevs,extend='max',cmap=get_cmap('turbo'))
axd['right'].set_xlabel('West-East (km)',fontsize=20,labelpad=6)
axd['right'].set_xticks(xax2)
axd['right'].set_xticklabels(xax1)
axd['right'].set_yticks(yax2)
axd['right'].set_yticklabels(['','','','','',''])
axd['right'].tick_params(axis='both', labelsize=20, which='major', pad=8,length=8)
axd['right'].set_aspect('equal')
axd['right'].grid()
axd['right'].text(0.0, 1.0, '(c) Low-level',transform=axd['right'].transAxes + trans,
            fontsize=20, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='k', pad=2))

### Set colorbar
cbar = fig.colorbar(im1, orientation="horizontal", cax=axd['cbar'])
cbar.ax.tick_params(labelsize=20, pad=6)

plt.savefig('/p/work1/lloveras/bwave_nov/figs_paper/cape_day5.pdf',bbox_inches='tight')
plt.show()
```
